```python
import pandas as pd
import numpy as np
print("modules imported successfully")
```

```python
myfile= open('one_piece.txt', 'r')
data= myfile.read()
words= data.split()
```

```python
def get_stats(ids):
    counts = {}
    for pair in zip(ids, ids[1:]): # Pythonic way to iterate consecutive elements
        counts[pair] = counts.get(pair, 0) + 1
    print(counts)

stats = get_stats(words)
```
